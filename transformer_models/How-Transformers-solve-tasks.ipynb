## How 🤗 Transformers solve tasks

- `Wav2Vec2` for audio classification and automatic speech recognition (ASR).
- `Vision Transformer (ViT)` and `ConvNeXT` for image classification.
- `DETR` for object detection.
- `Mask2Former` for image segmentation.
- `GLPN` for depth estimation.
- `BERT` for NLP tasks like text classification, token classification and question answering that use an encoder.
- `GPT2` for NLP tasks like text generation that use a decoder.
- `BART` for NLP tasks like summarization and translation that use an encoder-decoder.



### Transformer models for language

Language models are at the heart of modern NLP. They’re designed to understand and generate human language by learning the statistical patterns and relationships between words or tokens in text.

The Transformer was initially designed for machine translation, and since then, it has become the default architecture for solving all AI tasks. Some tasks lend themselves to the Transformer’s encoder structure, while others are better suited for the decoder. Still, other tasks make use of both the Transformer’s encoder-decoder structure.

### How language models work

Language models work by being trained to predict the probability of a word given the context of surrounding words. This gives them a foundational understanding of language that can generalize to other tasks.

There are two main approaches for training a transformer model:

- **Causal language modeling (CLM)**: The model is trained to predict the next word in a sequence, given the previous words. This is typically done using a decoder-only architecture, like GPT-2.

- **Masked language modeling (MLM)**: The model is trained to predict a masked word in a sequence, given the surrounding words. This is typically done using an encoder-only architecture, like BERT.


### Types of language models

In the Transformers library, language models generally fall into three architectural categories:

- **Encoder-only models**: These models, like `BERT`, are designed to understand the context of a given input. They are typically used for tasks like text classification, token classification, and question answering.

- **Decoder-only models**: These models, like `GPT-2`, `Llama`, are designed to generate text based on a given input. They are typically used for tasks like text generation and language modeling.

- **Encoder-decoder models**: These models, like `T5`, `BART`, combine both the encoder and decoder architectures. They are designed to take an input sequence and generate an output sequence, making them suitable for tasks like summarization and translation.

### Understanding the Transformer Architecture
it’s helpful to visualize how the **Encoder** and **Decoder** blocks interact. The core innovation of this architecture is the **Self-Attention** mechanism, which allows the model to weigh the importance of different words in a sequence regardless of their distance from each other.


### Key Components
*   **Multi-Head Attention:** This allows the model to jointly attend to information from different representation subspaces.
*   **Positional Encoding:** Since Transformers don't use `RNNs` or `CNNs`, they need a way to know the order of the words. This is added to the input embeddings.
*   **Feed-Forward Networks:** Each layer contains a fully connected feed-forward network applied to each position identically and independently.

The mathematical heart of the attention mechanism is defined by the formula:

$$ \text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V $$

Where:
*   $Q$ is the Query matrix.
*   $K$ is the Key matrix.
*   $V$ is the Value matrix.
*   $d_k$ is the dimension of the keys.
